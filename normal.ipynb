```python
from scipy.stats import binom, norm
import numpy as np
import matplotlib.pyplot as plt

# Parameters
N = 100       # Number of random numbers per experiment
experiments = 10000  # Number of experiments
bins = np.arange(0, N+2, 1)  # Bins for histogram with "drawers" of size 1

# Conduct experiments
sums = [np.sum(np.random.rand(N)) for _ in range(experiments)]

# Plot histogram of sums
plt.hist(sums, bins=bins, color='blue', alpha=0.5, label='Experimental')

# Normal approximation
mean = N / 2
variance = N / 12  # Variance for uniform distribution U(0,1) is (1/12)*((b-a)^2)
std_dev = np.sqrt(variance)
x = np.linspace(min(sums), max(sums), 1000)
plt.plot(x, norm.pdf(x, mean, std_dev) * experiments * (bins[1] - bins[0]), 'r-', label='Normal Approximation')

plt.title('Sum of 100 Random Numbers\n(Repeated 10,000 Times)')
plt.xlabel('Sum')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
